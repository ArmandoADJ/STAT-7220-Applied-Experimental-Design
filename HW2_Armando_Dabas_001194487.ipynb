```python
## Import Necessary Packages ##
from itertools import groupby
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd


## Import File##
path_file = "/content/STAT-7220-Applied-Experimental-Design/Assignments/HW2/Data Science Teaching Method.xlsx"
df = pd.read_excel(path_file)

##Print Columns names and first 6 rows#
df.head()

##Descriptive analysis of Teaching Methods##
print(df.groupby("Teaching_Method")["Score"].agg(["mean","std", "count"]))

##Data Visualization##
sns.boxplot(data = df, x = "Teaching_Method", y = "Score")
plt.title("Boxplot of scores grouped by teaching Methods")
plt.xlabel("Teaching Method")
plt.ylabel("Score")
plt.show()

print("\n")
sns.scatterplot(data = df, x = "Teaching_Method", y = "Score", hue = "Prior_Knowledge")
plt.title("Scatterplot of scores grouped by Teaching Methods")
plt.xlabel("Teaching Methods")
plt.ylabel("Score")
plt.show()

print("\n")

facet = sns.FacetGrid(df, col = "Teaching_Method", hue = "Teaching_Method")
facet.map(sns.histplot, "Score")
plt.show()

################################################################################


##Anova Test Model##
ols  = ols("Score ~ (Teaching_Method) + (Prior_Knowledge)", data = df).fit()
print(ols.summary())

###Plot Residuals###
##Check For Normality Assumption##

#Get Residuals#

ols_resid = ols.resid

#Plot Residuals qqplot#

fig = plt.figure(figsize= (10, 10))
ax = fig.add_subplot(111)

normality_plot, stat = stats.probplot(ols_resid, plot=plt, rvalue=True)
ax.set_title("Probability Plot of Model Residuals", fontsize=20)
ax.set

plt.show()

#Print Normality test(Shapiro test)#

print(stats.shapiro(ols_resid))

### Assessing Constant Variance ###

## Obtain Standardized Residuals ##

standardized_residuals = ols.get_influence().resid_studentized_internal

## Get the fitted values ##

fitted_values = ols.fittedvalues

## Create the plot ##

plt.figure(figsize=(8, 6))
plt.scatter(fitted_values, standardized_residuals)
plt.axhline(y=0, linestyle='dashed', color='red')
plt.axhline(y=3, color='blue')
plt.axhline(y=-3, color='blue')
plt.xlabel("Predicted y's")
plt.ylabel("Standardized Residuals")
plt.title("Standardized Residuals vs. Fitted Values")
plt.style.use('classic')
plt.show()

## B-P Test ##

## Import necessary function: ##

from statsmodels.stats.diagnostic import het_breuschpagan

## Run B-P Test ##

bp_test = het_breuschpagan(ols.resid,ols.model.exog)

## Extract F-statistic ##

f_statistic = bp_test[2]

## Extract its P-Value ##

f_p_value = bp_test[3]

print(f"F-statistic: {f_statistic}")
print(f"F p-value: {f_p_value}")

###Perform Tukey's HSD post-hoc Test###
tukey_result = pairwise_tukeyhsd(df["Score"], df["Teaching_Method"])
print(tukey_result)
```
